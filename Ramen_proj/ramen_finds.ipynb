```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ramen_df = pd.read_csv("ramen-ratings.csv", na_values="Unrated")
# first combine countries with different names for same country
ramen_df["Country"][ramen_df["Country"] == 'USA'] = 'United States'
```

```python
# sorted large to small
ramen_sorted = ramen_df.sort_values("Stars", ascending=False)
```

```python
ramen_df.head()
```

```python
ramen_sorted.head()
```

## The average amount of stars each type of style of ramen has

```python
sns.barplot(x=ramen_df["Style"], y=ramen_df["Stars"], ci=None)
```

> From this barplot we can assess that bar types of ramen are the best reviewed type of ramen based on this dataset.

## Now let's find out what country has the highest star ramen

```python
plt.figure(figsize=(8, 8), dpi=80, tight_layout=True)
sns.barplot(y="Country", x="Stars", data=ramen_df, ci=None, orient="h")
plt.show()
```

## Top brands of Ramen

```python
ramen_sorted = ramen_df.sort_values(["Stars", "Brand"], ascending=False)

plt.figure(figsize=(8, 50), dpi=80, tight_layout=True)
sns.barplot(y=ramen_df["Brand"], x=ramen_df["Stars"], ci=None, orient="h")
plt.show()
```
